```python
import pandas as pd
from pandas import json_normalize
import requests

# URL del JSON
url = "https://raw.githubusercontent.com/alura-cursos/challenge2-data-science-LATAM/refs/heads/main/TelecomX_Data.json"

# Descargar datos
response = requests.get(url)
response.raise_for_status()
data_json = response.json()

# Normalizar estructuras anidadas
df_full = json_normalize(data_json)

# Revisar estructura
print(df_full.shape)
df_full.info()
df_full.head()
```

```python
df_full.columns = (
    df_full.columns
    .str.lower()
    .str.replace(" ", "_")
    .str.replace(".", "_")
)
```

```python
#Reporte r치pido de calidad de datos

print("=== Dimensiones del DataFrame ===")
print(df_full.shape)
print()

print("=== Valores nulos por columna ===")
print(df_full.isna().sum().sort_values(ascending=False).head(10))
print()

print("=== Porcentaje de nulos ===")
print((df_full.isna().mean() * 100).round(2).sort_values(ascending=False).head(10))
print()

print("=== Duplicados ===")
print("N칰mero de filas duplicadas:", df_full.duplicated().sum())
print()

print("=== Valores 칰nicos en columnas categ칩ricas ===")
for col in df_full.select_dtypes(include="object").columns:
    print(f"{col}: {df_full[col].nunique()} valores 칰nicos")
    print(df_full[col].unique()[:10])  # muestra primeros 10 valores 칰nicos
    print("----")
```

```python
# Conteo de valores vac칤os en churn
empty_churn = (df_full['churn'] == '').sum()
total_rows = len(df_full)
percent_empty = empty_churn / total_rows * 100

print(f"Valores vac칤os en churn: {empty_churn}")
print(f"Porcentaje sobre la muestra: {percent_empty:.2f}%")
```

```python
import numpy as np

# Reemplazar valores vac칤os por NaN
df_full['churn'] = df_full['churn'].replace('', np.nan)

# Verificar resultado
print(df_full['churn'].value_counts(dropna=False))
```

```python
# Convertir cargos a num칠ricos
df_full['account_charges_total'] = pd.to_numeric(df_full['account_charges_total'], errors='coerce')
df_full['account_charges_monthly'] = pd.to_numeric(df_full['account_charges_monthly'], errors='coerce')

# Verificar tipos
print(df_full[['account_charges_total','account_charges_monthly']].dtypes)

# Convertir todas las columnas de texto a min칰sculas y eliminar espacios extra.
for col in df_full.select_dtypes(include="object").columns:
    df_full[col] = df_full[col].str.strip().str.lower()
```

```python
print(df_full[['account_charges_total','account_charges_monthly']].describe())
```

```python
# Crear columna cuentas_diarias
df_full['cuentas_diarias'] = df_full['account_charges_monthly'] / 30

# Verificar resultado
print(df_full[['account_charges_monthly', 'cuentas_diarias']].head())
```

```python
# Columnas binarias que solo tienen yes/no
binarias = [
    'churn',
    'customer_partner',
    'customer_dependents',
    'phone_phoneservice',
    'phone_multiplelines',
    'internet_onlinesecurity',
    'internet_onlinebackup',
    'internet_deviceprotection',
    'internet_techsupport',
    'internet_streamingtv',
    'internet_streamingmovies',
    'account_paperlessbilling'
]

for col in binarias:
    df_full[col] = df_full[col].map({'yes': 1, 'no': 0})
```

#游늵 Carga y an치lisis

```python
# Estad칤sticas descriptivas con formato claro
desc = df_full.describe()

# Convertir a object para evitar warnings
formatted = desc.astype(object)

# Columnas binarias
binarias = [
    'churn','customer_seniorcitizen','customer_partner','customer_dependents',
    'phone_phoneservice','phone_multiplelines','internet_onlinesecurity',
    'internet_onlinebackup','internet_deviceprotection','internet_techsupport',
    'internet_streamingtv','internet_streamingmovies','account_paperlessbilling'
]

for col in binarias:
    for idx in formatted.index:
        if idx == 'count':
            formatted.loc[idx, col] = f"{int(desc.loc[idx, col])}"
        else:
            formatted.loc[idx, col] = f"{desc.loc[idx, col]*100:.2f}%"

# Columnas monetarias
monetarias = ['account_charges_monthly','account_charges_total','cuentas_diarias']
for col in monetarias:
    for idx in formatted.index:
        if idx == 'count':
            formatted.loc[idx, col] = f"{int(desc.loc[idx, col])}"
        else:
            formatted.loc[idx, col] = f"${desc.loc[idx, col]:,.2f}"

# Tenure
col = 'customer_tenure'
for idx in formatted.index:
    if idx == 'count':
        formatted.loc[idx, col] = f"{int(desc.loc[idx, col])}"
    else:
        formatted.loc[idx, col] = f"{desc.loc[idx, col]:.2f}"

# Mostrar tabla
display(formatted)

# Notas interpretativas autom치ticas
print("\nNotas interpretativas:")
print(f"- churn: {desc.loc['mean','churn']*100:.2f}% de clientes se dieron de baja.")
print(f"- seniorcitizen: {desc.loc['mean','customer_seniorcitizen']*100:.2f}% son adultos mayores.")
print(f"- partner: {desc.loc['mean','customer_partner']*100:.2f}% tienen pareja.")
print(f"- dependents: {desc.loc['mean','customer_dependents']*100:.2f}% tienen dependientes.")
print(f"- tenure: mediana = {desc.loc['50%','customer_tenure']:.0f} meses, m치ximo = {desc.loc['max','customer_tenure']:.0f} meses.")
print(f"- cargos mensuales: media = ${desc.loc['mean','account_charges_monthly']:.2f}, mediana = ${desc.loc['50%','account_charges_monthly']:.2f}.")
print(f"- cargos totales: media = ${desc.loc['mean','account_charges_total']:.2f}, mediana = ${desc.loc['50%','account_charges_total']:.2f}, m치ximo = ${desc.loc['max','account_charges_total']:.2f}.")
print(f"- cuentas diarias: media = ${desc.loc['mean','cuentas_diarias']:.2f}, mediana = ${desc.loc['50%','cuentas_diarias']:.2f}.")
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Conteo de churn incluyendo NaN
churn_counts = df_full['churn'].value_counts(dropna=False)

# Reemplazar 칤ndices por etiquetas descriptivas
new_index = []
for val in churn_counts.index:
    if pd.isna(val):
        new_index.append('Dato faltante')
    elif val == 0:
        new_index.append('Fidelidad (0)')
    elif val == 1:
        new_index.append('Abandono (1)')
    else:
        new_index.append(val)

churn_counts.index = new_index

# Gr치fico de pastel
churn_counts.plot(
    kind='pie',
    autopct='%1.1f%%',
    colors=['skyblue','salmon','gray']
)
plt.title('Proporci칩n de churn')
plt.ylabel('')
plt.show()
```

```python
import pandas as pd

# Tablas de contingencia normalizadas (%)
tabla_genero = pd.crosstab(df_full['customer_gender'], df_full['churn'], normalize='index') * 100
tabla_contrato = pd.crosstab(df_full['account_contract'], df_full['churn'], normalize='index') * 100
tabla_pago = pd.crosstab(df_full['account_paymentmethod'], df_full['churn'], normalize='index') * 100

# Renombrar columnas de churn
tabla_genero = tabla_genero.rename(columns={0.0: 'Fidelidad (0)', 1.0: 'Abandono (1)'})
tabla_contrato = tabla_contrato.rename(columns={0.0: 'Fidelidad (0)', 1.0: 'Abandono (1)'})
tabla_pago = tabla_pago.rename(columns={0.0: 'Fidelidad (0)', 1.0: 'Abandono (1)'})

# Formatear a dos decimales con s칤mbolo %
tabla_genero = tabla_genero.astype(object).apply(lambda col: col.map(lambda x: f"{x:.2f}%"))
tabla_contrato = tabla_contrato.astype(object).apply(lambda col: col.map(lambda x: f"{x:.2f}%"))
tabla_pago = tabla_pago.astype(object).apply(lambda col: col.map(lambda x: f"{x:.2f}%"))

# Mostrar tablas
print("Churn seg칰n g칠nero:\n", tabla_genero, "\n")
print("Churn seg칰n contrato:\n", tabla_contrato, "\n")
print("Churn seg칰n m칠todo de pago:\n", tabla_pago, "\n")
```

```python
import pandas as pd

# Estad칤sticas de cargos totales y meses de servicio seg칰n churn
tabla_cargos = df_full.groupby('churn')['account_charges_total'].describe()
tabla_tenure = df_full.groupby('churn')['customer_tenure'].describe()

# Renombrar 칤ndices de churn
tabla_cargos.index = tabla_cargos.index.map({0.0: 'Fidelidad (0)', 1.0: 'Abandono (1)'})
tabla_tenure.index = tabla_tenure.index.map({0.0: 'Fidelidad (0)', 1.0: 'Abandono (1)'})

# Formatear cargos totales como moneda
tabla_cargos_fmt = tabla_cargos.astype(object)
for col in tabla_cargos.columns:
    if col == 'count':
        tabla_cargos_fmt[col] = tabla_cargos[col].astype(int)
    else:
        tabla_cargos_fmt[col] = tabla_cargos[col].map(lambda x: f"${x:,.2f}")

# Formatear tenure con dos decimales
tabla_tenure_fmt = tabla_tenure.astype(object)
for col in tabla_tenure.columns:
    if col == 'count':
        tabla_tenure_fmt[col] = tabla_tenure[col].astype(int)
    else:
        tabla_tenure_fmt[col] = tabla_tenure[col].map(lambda x: f"{x:.2f}")

# Mostrar tablas
print("Cargos totales seg칰n churn:\n", tabla_cargos_fmt, "\n")
print("Meses de servicio seg칰n churn:\n", tabla_tenure_fmt, "\n")

```

游꿢 Objetivo

El objetivo de este an치lisis fue identificar los factores asociados al abandono de clientes (churn) en Telecom X, mediante un proceso de ETL, limpieza de datos y an치lisis exploratorio, con el fin de proporcionar informaci칩n clara al equipo de ciencia de datos para el desarrollo de modelos predictivos.

游늷 1勇 Tasa General de Churn

El an치lisis muestra que el 26.54% de los clientes han abandonado el servicio.

Esto representa un nivel significativo de evasi칩n y confirma la necesidad de implementar estrategias de retenci칩n.

游눯 2勇 Cargos Totales y Abandono

Los clientes que permanecen activos presentan un mayor gasto acumulado.

Los clientes que abandonan muestran una mediana de cargos totales considerablemente menor.

游녤 Insight:
El churn se concentra en clientes con menor inversi칩n acumulada, lo que sugiere que el abandono ocurre en etapas tempranas del ciclo de vida del cliente.

낍 3勇 Antig칲edad del Cliente (Tenure)

Clientes fieles: mediana 곋 38 meses

Clientes que abandonan: mediana 곋 10 meses

游녤 Insight clave:
El abandono ocurre principalmente en clientes recientes.
La fidelizaci칩n durante los primeros meses es un factor cr칤tico.

游늼 4勇 Tipo de Contrato

Se observ칩 que:

Contratos mensuales presentan la mayor tasa de churn (42,71%)

Contratos anuales o bianuales presentan tasas significativamente menores.(11,27% y 2,83% respectivamente).

游녤 Insight:
Los contratos de mayor duraci칩n funcionan como mecanismo de retenci칩n.

游눱 5勇 M칠todo de Pago

El an치lisis indica que ciertos m칠todos de pago presentan mayor tasa de abandono. Los clientes con cheque electr칩nico tienen el churn m치s alto (45.29%), mientras que transferencias autom치ticas y tarjetas autom치ticas muestran menor abandono(16,71% y 15.24% respectivamente).
Esto podr칤a estar relacionado con menor compromiso o fricci칩n en el proceso de pago.


游댍 Conclusiones Estrat칠gicas

El churn en Telecom X est치 principalmente asociado a:

-Clientes con baja antig칲edad.

-Determinados m칠todos de pago.
